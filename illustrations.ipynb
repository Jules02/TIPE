```python
import matplotlib.pyplot as plt

import numpy as np

import cv2
import time
from skimage import color, transform
from skimage.feature import hog
import imutils
import os

import joblib

positive_patches_shape = (62, 47)

path_to_img = 'img/newsradio.jpg'

image = cv2.imread(path_to_img) 

image = color.rgb2gray(image)

(H_MAX, W_MAX) = (300, 600)

if image.shape[0] > H_MAX or image.shape[1] > W_MAX:
    s = min(H_MAX / image.shape[0], W_MAX / image.shape[1])
    image = transform.rescale(image, s)
    
#print("L'image étudiée est de taille {}".format(image.shape))

fig, ax = plt.subplots()
ax.imshow(image, cmap='gray')
ax.axis('off')

print(image)
```
